```python
import pandas as pd
```

```python
batting = 'https://www.baseball-reference.com/leagues/MLB/2020-standard-batting.shtml'
```

```python
tables = pd.read_html(batting)
batting_df = tables[0]
batting_df
```

```python
pitching = 'https://www.baseball-reference.com/leagues/MLB/2020-standard-pitching.shtml'
tables = pd.read_html(pitching)
pitching_df = tables[0]
pitching_df
```

```python
fielding = 'https://www.baseball-reference.com/leagues/MLB/2020-standard-fielding.shtml'
tables = pd.read_html(fielding)
fielding_df = tables[0]
fielding_df
```

```python
cols = [1,2,3,4,5,6]
batting_df.drop(batting_df.columns[cols],axis=1,inplace=True)
batting_df
```

```python
batting_df.drop(batting_df.iloc[:, 3:], inplace = True, axis = 1)
batting_df
```

```python
pitching_df.drop(pitching_df.iloc[:, 1:4], inplace = True, axis = 1)
pitching_df
```

```python
pitching_df.drop(pitching_df.iloc[:, 2:13], inplace = True, axis = 1)
pitching_df
```

```python
pitching_df.drop(pitching_df.iloc[:, 4:], inplace = True, axis = 1)
pitching_df
```

```python
fielding_df.drop(fielding_df.iloc[:, 1:11], inplace = True, axis = 1)
fielding_df
```

```python
fielding_df.drop(fielding_df.iloc[:, 2:], inplace = True, axis = 1)
fielding_df
```

```python
stat_df = pd.merge(pd.merge(batting_df,pitching_df,on='Tm'),fielding_df,on='Tm')
stat_df = stat_df.rename(columns={"R_x": "runs_for", "R_y": "runs_against",
                                  "H_x": "hits_for", "H_y": "hits_against",
                                 "Tm": "Team"})
stat_df
```

```python
stat_df = stat_df.drop([30,31,32])
stat_df
```

```python
stat_df.to_csv('stat_input_df.csv')
```
